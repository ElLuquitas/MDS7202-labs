```python
def plot_scatter(x, y, color):
    '''
    Entrega de manera gr√°fica un scatter plot de los datos entregados.

    Args:
        x (np.array): Coordenadas en x de los puntos a graficar (observaciones).
        y (np.array): Coordenadas en y de los puntos a graficar (clases).
        color (np.array): Clases a graficar diferenciadas por colores.

    Returns:
        fig (plotly.graph_objects.Figure): Gr√°fico de puntos.
    '''
    fig = go.Scatter(
        x = x,
        y = y,
        mode = 'markers',
        marker = dict(
            color = color,
            colorscale = 'Viridis',
            showscale = False
        )
    )
    return fig

def plot_multi_scatter(matrix_data_sets):
    '''
    Entrega de manera gr√°fica un scatter plot de los datos ya clusterizados y
    que vienen dentro de una matriz.

    Args:
        matrix_data_sets (list): Lista con los datos a graficar, ordenados como matriz.
    '''
    fig = make_subplots(
        rows = 3,
        cols = 4,
        subplot_titles = ['Kmeans', 'GMM', 'WARD', 'DBSCAN']
    )

    # Iteraci√≥n sobre los datasets
    for i, (data, exe_time, silhouette_score) in enumerate(matrix_data_sets):
        fig.add_trace(
            plot_scatter(data['x'], data['y'], data['pred']),
            row = i // 4 + 1, col = i % 4 + 1
        )

        # Agregar el tiempo y el silhouette en la esquina superior derecha
        fig.add_annotation(
            text=f"{exe_time:.2f} [s] | s: {silhouette_score:.2f}",
            xref="x domain",
            yref="y domain",
            x=1, y=1,
            xanchor="right",
            yanchor="top",
            showarrow=False,
            font=dict(size=10),
            row=i // 4 + 1,
            col=i % 4 + 1

            
        )

    fig.update_layout(height=800, width=1000, title_text="Comparaci√≥n de m√©todos de clustering", showlegend=False)
    
    fig.show()
```

```python
#Se crean los tres dataset
data_sets_1 = create_data(1000)
data_sets_2 = create_data(5000)
data_sets_3 = create_data(10000)
```

```python
# Aplicar algoritmos de clustering sobre los datos ficticios
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture 
from sklearn.metrics import silhouette_score
import time

# Funci√≥n que se encarga del proceso de clustering
def cluster_model_apply(data_set, model, random_state = 30):
    '''
    Aplica un algoritmo de clustering sobre un dataset y devuelve el resultado
    en un DataFrame, el tiempo de ejecuci√≥n y el silhouette score.

    Args:
        data_set (dict): Diccionario con los datos a clusterizar.
        model (sklearn.base.BaseEstimator): Algoritmo de clustering a aplicar.

    Returns:
        df (pd.DataFrame): DataFrame con las predicciones del modelo.
        exe_time (float): Tiempo de ejecuci√≥n del algoritmo.
        silhouette_score (float): Valor del silhouette score.
    '''
    # Extraer datos

    x = data_set['x']
    classes = data_set['classes']
    n_cluster = data_set['n_cluster']

    # Aplicaci√≥n de modelo y configuraci√≥n dependiendo de cu√°l sea
    model_instance = model()
    if model == KMeans:
        model_instance = model_instance.set_params(n_clusters = n_cluster, random_state = random_state)

    elif model == GaussianMixture:
        model_instance = model_instance.set_params(n_components = n_cluster, random_state = random_state)

    elif model == AgglomerativeClustering:
        model_instance = model_instance.set_params(n_clusters = n_cluster, linkage = 'ward')

    elif model == DBSCAN:
        model_instance = model_instance.set_params(eps = 0.4, min_samples = 3)
    
    init = time.time()
    pred = model_instance.fit_predict(x)
    end = time.time()

    # Tiempo de ejecuci√≥n
    exe_time = (end - init).__round__(2)

    # Calcular silhouette score (excepci√≥n creada para DBSCAN que fallaba)
    if len(set(pred)) > 1:
        silhouette = silhouette_score(x, pred).__round__(2)
    else:
        silhouette = float('nan')

    # Crear DataFrame
    df = pd.DataFrame({
        'x': x[:, 0],
        'y': x[:, 1],
        'class': classes,
        'pred': pred
    })

    return df, exe_time, silhouette
```

```python
# Aplicaci√≥n de Kmeans
data_sets_1_kmeans = [cluster_model_apply(data_sets_1[key], KMeans) for key in data_sets_1]
data_sets_2_kmeans = [cluster_model_apply(data_sets_2[key], KMeans) for key in data_sets_2]
data_sets_3_kmeans = [cluster_model_apply(data_sets_3[key], KMeans) for key in data_sets_3]

# Aplicaci√≥n de GMM
data_sets_1_gmm = [cluster_model_apply(data_sets_1[key], GaussianMixture) for key in data_sets_1]
data_sets_2_gmm = [cluster_model_apply(data_sets_2[key], GaussianMixture) for key in data_sets_2]
data_sets_3_gmm = [cluster_model_apply(data_sets_3[key], GaussianMixture) for key in data_sets_3]

# Aplicaci√≥n de Ward
data_sets_1_ward = [cluster_model_apply(data_sets_1[key], AgglomerativeClustering) for key in data_sets_1]
data_sets_2_ward = [cluster_model_apply(data_sets_2[key], AgglomerativeClustering) for key in data_sets_2]
data_sets_3_ward = [cluster_model_apply(data_sets_3[key], AgglomerativeClustering) for key in data_sets_3]

# Aplicaci√≥n de DBSCAN
data_sets_1_dbscan = [cluster_model_apply(data_sets_1[key], DBSCAN) for key in data_sets_1]
data_sets_2_dbscan = [cluster_model_apply(data_sets_2[key], DBSCAN) for key in data_sets_2]
data_sets_3_dbscan = [cluster_model_apply(data_sets_3[key], DBSCAN) for key in data_sets_3]
```

**2. Explique qu√© efectos podr√≠a causar el uso de variables categ√≥ricas en un algoritmo no supervisado.**
El uso de variables categ√≥ricas sin tener un preprocesamiento en algoritmos no supervisados, como clustering, puede generar varios problemas:

1. **C√°lculo de distancias no aplicable**: Los algoritmos basados en distancias, como K-Means, no pueden calcular distancias entre variables categ√≥ricas, lo que genera agrupaciones incorrectas cuando se usan variable categ√≥rica. 

2. **Escalado desbalanceado**: Cuando convertimos las variables categ√≥ricas en varias columnas usando One-Hot Encoding (esta es una t√©cnica para transformar variables categ√≥ricas), √©stas pueden tener m√°s influencia de lo que deber√≠an en los resultados del agrupamiento, lo que puede distorsionar el an√°lisis.

3. **Agrupaciones no significativas**: Las categor√≠as pueden ser interpretadas como n√∫meros arbitrarios, lo que genera grupos sin sentido.

4. **P√©rdida de informaci√≥n sem√°ntica**: Sin preprocesamiento adecuado, se pierde el significado importante de las categor√≠as, afectando la calidad de los resultados.

2. Explique cu√°l ser√≠a el criterio adecuado para seleccionar el n√∫mero √≥ptimo de cl√∫sters. **Justifique de forma estadistica y a traves de gr√°ficos.** [6 puntos]

> **HINT:** Se recomienda investigar sobre los criterios AIC y BIC para esta tarea.

**Respuestas:**

```python
def evaluate_gmm_aic_bic(df, variables, n_clusters_range):
    """
    Eval√∫a un modelo de mezcla gaussiana para diferentes n√∫meros de cl√∫sters 
    y grafica las puntuaciones AIC y BIC.

    Par√°metros:
        df (DataFrame): DataFrame que contiene los datos.
        variables (list): Lista de nombres de las columnas a usar.
        n_clusters_range (range): Rango de n√∫meros de cl√∫sters a explorar.
    """
    # Definir el pipeline
    pipeline = Pipeline([
        ('gmm', GaussianMixture(random_state=42))
    ])

    # Listas para almacenar los resultados de AIC y BIC
    aic_scores = []
    bic_scores = []

    # Iterar sobre los diferentes n√∫meros de cl√∫sters
    for n_clusters in n_clusters_range:
        # Ajustar el modelo con el n√∫mero de cl√∫sters actual
        pipeline.set_params(gmm__n_components=n_clusters)
        pipeline.fit(df[variables])  # Usar solo las columnas seleccionadas
        
        # Obtener las m√©tricas AIC y BIC
        gmm_model = pipeline.named_steps['gmm']
        aic_scores.append(gmm_model.aic(df[variables]))  # Usar las mismas columnas
        bic_scores.append(gmm_model.bic(df[variables]))  # Usar las mismas columnas

    # Graficar los resultados de AIC y BIC
    plt.figure(figsize=(10, 6))
    plt.plot(n_clusters_range, aic_scores, label='AIC', marker='o')
    plt.plot(n_clusters_range, bic_scores, label='BIC', marker='o')
    plt.title('Comparaci√≥n de AIC y BIC para seleccionar el n√∫mero de cl√∫sters')
    plt.xlabel('N√∫mero de cl√∫sters')
    plt.ylabel('Puntuaci√≥n')
    plt.legend()
    plt.show()

```

```python
# Uso de la funci√≥n
variables_usadas = ['Inflight wifi service', 'Inflight entertainment', 'Cleanliness', 'Departure Delay in Minutes']
n_clusters_range = range(3, 9)

evaluate_gmm_aic_bic(df_transformed, variables_usadas, n_clusters_range)
```

El AIC (Criterio de Informaci√≥n de Akaike) y BIC (Criterio de Informaci√≥n Bayesiano) son criterios para seleccionar n√∫meros de clusters. Las m√©tricas est√°n hechas para comparar, por lo que m√°s que el valor, lo que nos sirve es la diferencia entre las distintas opciones que generan AIC y BIC entre 3 y 8 clusters. **Respuestas:**

```python
def generate_clustered_df(df, variables, n_clusters):
    """
    Genera un DataFrame que incluye solo las variables seleccionadas, las etiquetas de cl√∫ster,
    y las primeras dos componentes principales (PC1 y PC2).
    
    Par√°metros:
        df (DataFrame): DataFrame que contiene los datos.
        variables (list): Lista de nombres de las columnas a usar.
        n_clusters (int): N√∫mero de cl√∫sters elegidos.
    
    Retorna:
        DataFrame: DataFrame con las variables seleccionadas, los cl√∫sters, PC1 y PC2.
    """
    # Definir el pipeline
    pipeline = Pipeline([
        ('gmm', GaussianMixture(n_components=n_clusters, random_state=42))
    ])
    
    # Ajustar el modelo
    pipeline.fit(df[variables])
    
    # Predecir los cl√∫sters
    labels = pipeline.predict(df[variables])
    
    # Aplicar PCA para obtener las primeras dos componentes principales
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(df[variables])
    
    # Crear un DataFrame con las variables seleccionadas
    selected_vars_df = df[variables].copy()
    
    # Crear un DataFrame con los componentes principales
    pc_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])
    
    # Agregar las etiquetas de cl√∫ster y las componentes principales al DataFrame
    selected_vars_df['Cluster'] = labels
    selected_vars_df = pd.concat([selected_vars_df, pc_df], axis=1)
    
    return selected_vars_df
```

```python
# Uso de la funci√≥n:
variables_usadas = ['Inflight wifi service', 'Inflight entertainment', 'Cleanliness', 'Departure Delay in Minutes']
n_clusters_elegido = 6  # Cambia este valor seg√∫n lo que elijas

# Generar el DataFrame con cl√∫steres y componentes principales
df_clustered = generate_clustered_df(df_transformed, variables_usadas, n_clusters_elegido)

# Mostrar el DataFrame resultante
print(df_clustered.head())
```

```python
def plot_pca_clusters(df_clustered):
    """
    Grafica las primeras dos componentes principales (PC1 y PC2) y colorea los puntos seg√∫n el cl√∫ster.

    Par√°metros:
        df_clustered (DataFrame): DataFrame que contiene las columnas 'PC1', 'PC2', y 'Cluster'.
    """
    # Graficar las componentes principales con el color basado en los cl√∫sters
    fig = px.scatter(
        df_clustered, x='PC1', y='PC2', color='Cluster', 
        title='Visualizaci√≥n de Cl√∫sters usando PCA',
        labels={'PC1': 'Componente Principal 1', 'PC2': 'Componente Principal 2'},
        color_continuous_scale=px.colors.qualitative.Set1  # Colores separados en lugar de un gradiente
    )
    
    # Mostrar el gr√°fico
    fig.show()
```

```python
#3. Proporcionar una descripci√≥n breve de cada cl√∫ster utilizando estad√≠sticas descriptivas b√°sicas, 
#como la media y la desviaci√≥n est√°ndar, para resumir las caracter√≠sticas de las variables utilizadas en estos algoritmos

def estadistica(df_clustered, variables):
    """
    Calcula estad√≠sticas descriptivas combinadas (media, desviaci√≥n est√°ndar, min, max, cuartiles)
    para cada cl√∫ster considerando todas las variables seleccionadas.

    Par√°metros:
    df_clustered (DataFrame): DataFrame que contiene los datos con una columna 'Cluster'.
    variables (list): Lista de nombres de las columnas a usar para el an√°lisis.
    """
    # Agrupar por cl√∫ster
    clusters = df_clustered.groupby('Cluster')

    # Crear una lista para almacenar los resultados
    results = []

    # Calcular estad√≠sticas combinadas para cada cl√∫ster
    for cluster, data in clusters:
        # Subconjunto del dataframe con las variables seleccionadas
        subset = data[variables]

        # Verificar que no todos los valores sean NaN
        if not subset.isna().all().all():
            # Calcular las estad√≠sticas combinadas
            mean_value = subset.mean().mean()  # Media combinada de todas las variables
            std_value = subset.stack().std()   # Desviaci√≥n est√°ndar combinada
            min_value = subset.min().min()     # M√≠nimo de todas las variables
            q25_value = subset.stack().quantile(0.25)  # Cuartil 25%
            q50_value = subset.stack().quantile(0.5)   # Mediana (cuartil 50%)
            q75_value = subset.stack().quantile(0.75)  # Cuartil 75%
            max_value = subset.max().max()     # M√°ximo de todas las variables

            # Agregar los resultados a la lista
            results.append({
                'Cluster': cluster,
                'Mean': mean_value,
                'Std': std_value,
                'Min': min_value,
                '25%': q25_value,
                '50%': q50_value,
                '75%': q75_value,
                'Max': max_value
            })

    # Convertir la lista de resultados a un DataFrame
    cluster_stats = pd.DataFrame(results)

    # Mostrar las estad√≠sticas combinadas por cl√∫ster
    print(cluster_stats)
```

El 25% de los datos se encuentran muy cerca de 0, pero la mediana sugiere que la mayor√≠a de los datos est√°n por encima de 0.4.

```python
#4.Proceda a visualizar los cl√∫sters en tres dimensiones para una perspectiva m√°s detallada.
import pandas as pd
from sklearn.decomposition import PCA

def generate_pca(df, variables, n_components=3):
    """
    Genera componentes principales utilizando PCA y agrega al DataFrame original.

    Par√°metros:
        df (DataFrame): DataFrame que contiene los datos y los cl√∫steres.
        variables (list): Lista de nombres de las columnas a usar.
        n_components (int): N√∫mero de componentes principales a generar.

    Retorna:
        DataFrame: DataFrame con las componentes principales y las etiquetas de los cl√∫steres.
    """
    # Ajustar PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(df[variables])
    
    # Crear un DataFrame para las componentes principales
    pca_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])
    
    # Agregar las etiquetas de cl√∫ster al nuevo DataFrame
    pca_df['Cluster'] = df['Cluster'].values
    
    return pca_df

def plot_3d_pca(pca_df):
    """
    Grafica las componentes principales en 3D, coloreadas por los cl√∫steres.

    Par√°metros:
        pca_df (DataFrame): DataFrame que contiene las componentes principales y las etiquetas de los cl√∫steres.
    """
    # Crear una lista de colores separados
    unique_clusters = pca_df['Cluster'].unique()
    colors = px.colors.qualitative.Set1[:len(unique_clusters)]  # Selecciona colores de la paleta

    fig = px.scatter_3d(
        pca_df, 
        x='PC1', 
        y='PC2', 
        z='PC3', 
        color='Cluster',
        color_discrete_sequence=colors,  # Usar colores separados
        title='Componentes Principales en 3D',
        labels={'PC1': 'Componente Principal 1', 'PC2': 'Componente Principal 2', 'PC3': 'Componente Principal 3'}
    )
    fig.update_traces(marker=dict(size=5))
    fig.show()


```

```python
# Variables seleccionadas para PCA
variables_usadas = ['Inflight wifi service', 'Inflight entertainment', 'Cleanliness', 'Departure Delay in Minutes']

# Generar el DataFrame con las componentes principales
pca_df = generate_pca(df_clustered, variables_usadas)

# Graficar las componentes principales
plot_3d_pca(pca_df)
```
