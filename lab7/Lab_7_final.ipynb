1.  **Accuracy (Exactitud)**:fracci贸n de las predicciones correctas con el total de predicciones:

$$Accuracy = \frac{\text{Verdadero positivo+ verdadero negativo}}{\text{n煤mero de predicciones totales}}= \frac{TP+TN}{(TP+TN+FP+FN)}$$

  **Interpretaci贸n**:
  La exactitud mide el porcentaje de predicciones correctas entre todas las predicciones hechas. Se interpreta como el porcentaje de casos en los que el modelo acert贸, tanto para las clases positivas como negativas.

2. **Precision**: Fracci贸n de los verdaderos positivos con la suma de verdaderos y falsos positivos:

$$P = \frac{\text{Verdadero positivo}}{\text{Verdaderos positivos + falsos positivos}}  = \frac{TP}{(TP+FP)}$$

   **Interpretaci贸n**:
La precisi贸n mide el porcentaje de verdaderos positivos entre todas las predicciones que el modelo clasific贸 como positivas. Se interpreta como la capacidad del modelo de evitar falsos positivos. Es especialmente importante en situaciones donde los falsos positivos no son deseados, como en la detecci贸n  diagn贸sticos m茅dicos equivocados.

3. **Recall**: Fracci贸n de ejemplos `+` que son correctamente clasificados:

$$R = \frac{\text{Clasificados correctamente como positivo}}{\text{Todos los que deber铆a haber clasificado como positivos}}  = \frac{TP}{(TP+FN)}$$

   **Interpretaci贸n**:
El recall mide el porcentaje de verdaderos positivos que el modelo fue capaz de identificar correctamente, entre todos los casos que realmente eran positivos. Se interpreta como la capacidad del modelo para encontrar todos los casos positivos.


4. **F1 score**: mezcla la precisi贸n y recall usando una media arm贸nica.

$$F = \frac{2PR}{(P+R)}$$

   El F1 score es la media arm贸nica entre la precisi贸n y el recall, y se utiliza cuando es necesario un equilibrio entre ambos. Es una m茅trica 煤til cuando hay un desequilibrio entre clases o cuando tanto los falsos positivos como los falsos negativos no son deseados por las consecuencias que traen (por ejemplos, malos diagnosticos m茅dicos). Un alto F1 score indica que el modelo tiene tanto buena precisi贸n como buen recall, lo que lo hace adecuado para situaciones donde ambos errores (falsos positivos y falsos negativos) deben minimizarse.

Tenemos diferentes tipos de datos en este dataset, existiendo caracter铆sticas num茅ricas que tienen poca variaci贸n (por ejemplo `National_Positions`) y otras con bastante variaci贸n (`Dribbling`, `Aggression`, entre otras), adem谩s de variables de tipo *string* que pueden tener muchos o pocos valores. En resumen, tenemos los siguientes tipos de categor铆as:

1. **Variables Num茅ricas**: `Height`, `Weight`, `Age`, `Ball_Control`, `Dribbling`, `Marking`, `Sliding_Tackle`, `Standing_Tackle`, `Aggression`, `Reactions`, `Interceptions`, `Vision`, `Composure`, `Crossing`, `Short_Pass`, `Long_Pass`, `Acceleration`, `Speed`, `Stamina`, `Strength`, `Balance`, `Agility`, `Jumping`, `Heading`, `Shot_Power`, `Finishing`, `Long_Shots`, `Curve`, `Freekick_Accuracy`, `Penalties`, `Volleys`.
2. **Variables Categ贸ricas**: `Nationality`, `National_Position`, `Club_Position`, `Preffered_Foot`, `Work_Rate`, `Weak_foot`, `Skill_Moves`.
3. **Variables tipo string**: `Name`.

Esto nos lleva a pensar de que podemos hacer las siguientes transformaciones:

````python
# Transformar las variables num茅ricas tipo calificaci贸n con MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

num_columns = ['Height', 'Weight', 'Age', 'Ball_Control', 'Dribbling', 'Marking', 'Sliding_Tackle', 'Standing_Tackle',
               'Aggression', 'Reactions', 'Interceptions', 'Vision', 'Composure', 'Crossing', 'Short_Pass', 'Long_Pass',
               'Acceleration', 'Speed', 'Stamina', 'Strength', 'Balance', 'Agility', 'Jumping', 'Heading', 'Shot_Power',
               'Finishing', 'Long_Shots', 'Curve', 'Freekick_Accuracy', 'Penalties', 'Volleys']
num_scaler = MinMaxScaler()

# Transformar las variables categ贸ricas con OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

cat_columns = ['Club_Position', 'Preffered_Foot', 'Work_Rate', 'Weak_foot', 'Skill_Moves']
cat_encoder = OneHotEncoder(handle_unknown='infrequent_if_exist')

# Crear un ColumnTransformer para aplicar las transformaciones a las columnas correspondientes
from sklearn.compose import ColumnTransformer

col_transformer = ColumnTransformer([
    ('num', num_scaler, num_columns),
    ('cat', cat_encoder, cat_columns)
])
````

4. Justificaci贸n

Estas transformaciones se realizan por las siguientes razones:

* Dentro de las variables num茅ricas existen $3$ que podr铆amos considerar como una excepci贸n a la generalizaci贸n que se va a explicar: `Height`, `Weight` y `Age`. El resto de variables (`Dribbling`, `Aggression`, etc.) son una calificaci贸n que se le da al jugador entre $1$ y $100$ (aunque no existen jugadores con alguna marca perfecta). Todas estas variables num茅ricas las podemos normalizar entre $0$ y $1$ usando `MinMaxScaler`.

* Las variables `Height`, `Weight` y `Age` no se rankean entre $1$ y $100$ y no significan un "m谩s alto es mejor". De todas formas se deben escalar a valores entre $0$ y $1$.

* Las variables categ贸ricas justamente categorizan a los jugadores seg煤n procedencia y otros aspectos m谩s cualitativos, adem谩s de existir variables de tipo `string` (nacionalidades y posiciones). Las variables cuantitativas siguen siendo una categorizaci贸n con un orden espec铆fico. Para todas estas se usa una transformaci贸n tipo `OneHotEncoder`.

Adem谩s, no trabajaremos con las siguientes variables:

* `Nationality`: Puede sesgar a jugadores seg煤n su pa铆s de procedencia.

**Respuesta:**

1. Separaci贸n de datos

````python
from sklearn.model_selection import train_test_split

features = df.drop(columns = ['Name', 'Nationality', 'National_Position', 'label'])
labels = df.loc[:, 'label']

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.33, shuffle=True, stratify=labels, random_state=30
)
````

````python
# Revisar la distribuci贸n de etiquetas en los conjuntos de entrenamiento y prueba
print('Distribuci贸n de etiquetas en el conjunto de entrenamiento:')
y_train.value_counts() / y_train.count() * 100
````

````python
print('Distribuci贸n de etiquetas en el conjunto de prueba:')
y_test.value_counts() / y_test.count() * 100
````

2. Definici贸n de *pipeline*

````python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.pipeline import Pipeline

# Definici贸n de pipeline para entrenamiento con XGBoost
xgb_model = XGBClassifier()
pipeline_xgboost = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', xgb_model)
], verbose=True)

# Definici贸n de pipeline para entrenamiento con LightGBM
lgb_model = LGBMClassifier(verbose=-1)
pipeline_lightgbm = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', lgb_model)
], verbose=True)
````

3. Entrenamiento del *pipeline*

````python
# Entrenamiento de XGBoost
pipeline_xgboost.fit(X_train, y_train)
````

````python
# Entrenamiento de LightGBM
pipeline_lightgbm.fit(X_train, y_train)
````

* Precision: Calcula, en base a todas las observaciones categorizadas como clase positiva, la proporci贸n de observaciones correctamente clasificadas. Representado como f贸rmula:

\begin{equation*}
    precision = \frac{TP}{TP + FP}

Esto nos indica que, dentro de las observaciones catalogadas como positivas, aproximadamente un poco m谩s de la mitad correspond铆an a la clase verdadera.

* Recall: Calcula, en base a todas las observaciones que deb铆an ser clase positiva, la proporci贸n de observaciones correctamente clasificadas. Representado como f贸rmula:

\begin{equation*}
    recall = \frac{TP}{TP + FN}
\end{equation*}

Por otro lado, esta m茅trica obtiene resultados muy bajos, ya que opera sobre las observaciones que pertenecen a la clase positiva, viendo que existe un claro sesgo hacia la clase negativa, ya que muchas observaciones positivas se predijeron err贸neamente, consiguiendo $recall = 0.13$ para `XGBoost` y $recall = 0.12$ para `LightGBM`.

A modo general, ambos modelos obtuvieron excelentes m茅tricas en relaci贸n a la clase negativa.
