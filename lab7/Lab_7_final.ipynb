La matriz de confusión se usa para evaluar el rendimiento de un modelo de clasificación. En clase, vimos que es una tabla que resume las predicciones del modelo en comparación con las etiquetas verdaderas, mostrando qué tan bien ha clasificado correctamente los datos y dónde ha cometido errores. La matriz de confusión organiza los resultados en forma de una cuadrícula que permite identificar verdaderos positivos (TP), falsos negativos (FN), falsos positivos (FP) y verdaderos negativos (TN). Estas cuatro categorías son importantes de entender: el verdadero positivo es cuando el modelo predijo correctamente la clase positiva, el verdadero negativo es cuando el modelo predijo correctamente la clase negativa, el falso positivo es cuando el modelo predijo erróneamente que era clase positiva pero en realidad era negativa, y el falso negativo es cuando predijo negativa, pero en realidad era positiva. Un ejemplo claro es el caso de las pruebas de COVID, donde a veces las personas estaban enfermas, pero la prueba no daba positivo, o al revés, las personas no estaban enfermas y la prueba daba positivo.

**Pregunta 4**

1.  **Accuracy (Exactitud)**:fracción de las predicciones correctas con el total de predicciones:

$$Accuracy = \frac{\text{Verdadero positivo+ verdadero negativo}}{\text{número de predicciones totales}}= \frac{TP+TN}{(TP+TN+FP+FN)}$$

  **Interpretación**:
  La exactitud mide el porcentaje de predicciones correctas entre todas las predicciones hechas. Se interpreta como el porcentaje de casos en los que el modelo acertó, tanto para las clases positivas como negativas.

2. **Precision**: Fracción de los verdaderos positivos con la suma de verdaderos y falsos positivos:

$$P = \frac{\text{Verdadero positivo}}{\text{Verdaderos positivos + falsos positivos}}  = \frac{TP}{(TP+FP)}$$

   **Interpretación**:
La precisión mide el porcentaje de verdaderos positivos entre todas las predicciones que el modelo clasificó como positivas. Se interpreta como la capacidad del modelo de evitar falsos positivos. Es especialmente importante en situaciones donde los falsos positivos no son deseados, como en la detección  diagnósticos médicos equivocados.

3. **Recall**: Fracción de ejemplos `+` que son correctamente clasificados:

$$R = \frac{\text{Clasificados correctamente como positivo}}{\text{Todos los que debería haber clasificado como positivos}}  = \frac{TP}{(TP+FN)}$$

   **Interpretación**:
El recall mide el porcentaje de verdaderos positivos que el modelo fue capaz de identificar correctamente, entre todos los casos que realmente eran positivos. Se interpreta como la capacidad del modelo para encontrar todos los casos positivos.


4. **F1 score**: mezcla la precisión y recall usando una media armónica.

$$F = \frac{2PR}{(P+R)}$$

   El F1 score es la media armónica entre la precisión y el recall, y se utiliza cuando es necesario un equilibrio entre ambos. Es una métrica útil cuando hay un desequilibrio entre clases o cuando tanto los falsos positivos como los falsos negativos no son deseados por las consecuencias que traen (por ejemplos, malos diagnosticos médicos). Un alto F1 score indica que el modelo tiene tanto buena precisión como buen recall, lo que lo hace adecuado para situaciones donde ambos errores (falsos positivos y falsos negativos) deben minimizarse.

**Pregunta 5**

A continuación se eligen las métricas más adecuadas según el caso planteado:
   
1. **Mantenimiento Predictivo**: Recall. Se elige Recall ya que en el mantenimiento predictivo, es importante identificar todas las posibles fallas antes de que ocurran, para evitar paradas no planificadas pérdidas de producción. Un alto recall asegura que la mayoría de las fallas reales sean detectadas por el modelo. Aunque un recall alto puede generar algunos falsos positivos (alertas innecesarias), en el contexto de mantenimiento predictivo, es más aceptable tener algunas alertas adicionales que el riesgo de no detectar una falla crítica.

2. **Detección de Enfermedades Contagiosas**: Recall. En la detección de enfermedades contagiosas, es esencial identificar a todas las personas infectadas para prevenir la propagación de la enfermedad. Un alto recall asegura que la mayoría de los casos positivos sean detectados, aunque hayan algunos falsos negativos, ya que lo vital es evitar la propagación de la enfermedad.  

3. **Aprobación de Créditos de Alto Riesgo**: Precision. En la aprobación de créditos de alto riesgo, es importante minimizar los falsos positivos, es decir, evitar aprobar créditos a personas que probablemente no paguen (riesgo de deudores) que resulte en pérdidas financieras.

4. **Detección de Crímenes**: F1 Score. La detección de crímenes requiere un equilibrio entre la precisión y el recall. Se necesita identificar correctamente la mayor cantidad de crímenes posibles (alto recall) sin generar demasiados falsos positivos que podrían llevar a acusaciones injustas, malgasto de recursos o personas inocentes culpables.

Tenemos diferentes tipos de datos en este dataset, existiendo características numéricas que tienen poca variación (por ejemplo `National_Positions`) y otras con bastante variación (`Dribbling`, `Aggression`, entre otras), además de variables de tipo *string* que pueden tener muchos o pocos valores. En resumen, tenemos los siguientes tipos de categorías:

1. **Variables Numéricas**: `Height`, `Weight`, `Age`, `Ball_Control`, `Dribbling`, `Marking`, `Sliding_Tackle`, `Standing_Tackle`, `Aggression`, `Reactions`, `Interceptions`, `Vision`, `Composure`, `Crossing`, `Short_Pass`, `Long_Pass`, `Acceleration`, `Speed`, `Stamina`, `Strength`, `Balance`, `Agility`, `Jumping`, `Heading`, `Shot_Power`, `Finishing`, `Long_Shots`, `Curve`, `Freekick_Accuracy`, `Penalties`, `Volleys`.
2. **Variables Categóricas**: `Nationality`, `National_Position`, `Club_Position`, `Preffered_Foot`, `Work_Rate`, `Weak_foot`, `Skill_Moves`.
3. **Variables tipo string**: `Name`.

Esto nos lleva a pensar de que podemos hacer las siguientes transformaciones:

````python
# Transformar las variables numéricas tipo calificación con MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

num_columns = ['Height', 'Weight', 'Age', 'Ball_Control', 'Dribbling', 'Marking', 'Sliding_Tackle', 'Standing_Tackle',
               'Aggression', 'Reactions', 'Interceptions', 'Vision', 'Composure', 'Crossing', 'Short_Pass', 'Long_Pass',
               'Acceleration', 'Speed', 'Stamina', 'Strength', 'Balance', 'Agility', 'Jumping', 'Heading', 'Shot_Power',
               'Finishing', 'Long_Shots', 'Curve', 'Freekick_Accuracy', 'Penalties', 'Volleys']
num_scaler = MinMaxScaler()

# Transformar las variables categóricas con OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

cat_columns = ['Club_Position', 'Preffered_Foot', 'Work_Rate', 'Weak_foot', 'Skill_Moves']
cat_encoder = OneHotEncoder(handle_unknown='infrequent_if_exist')

# Crear un ColumnTransformer para aplicar las transformaciones a las columnas correspondientes
from sklearn.compose import ColumnTransformer

col_transformer = ColumnTransformer([
    ('num', num_scaler, num_columns),
    ('cat', cat_encoder, cat_columns)
])
````

4. Justificación

Estas transformaciones se realizan por las siguientes razones:

* Dentro de las variables numéricas existen $3$ que podríamos considerar como una excepción a la generalización que se va a explicar: `Height`, `Weight` y `Age`. El resto de variables (`Dribbling`, `Aggression`, etc.) son una calificación que se le da al jugador entre $1$ y $100$ (aunque no existen jugadores con alguna marca perfecta). Todas estas variables numéricas las podemos normalizar entre $0$ y $1$ usando `MinMaxScaler`.

* Las variables `Height`, `Weight` y `Age` no se rankean entre $1$ y $100$ y no significan un "más alto es mejor". De todas formas se deben escalar a valores entre $0$ y $1$.

* Las variables categóricas justamente categorizan a los jugadores según procedencia y otros aspectos más cualitativos, además de existir variables de tipo `string` (nacionalidades y posiciones). Las variables cuantitativas siguen siendo una categorización con un orden específico. Para todas estas se usa una transformación tipo `OneHotEncoder`.

Además, no trabajaremos con las siguientes variables:

* `Name`: Identifica simplemente al jugador.
* `Nationality`: Puede sesgar a jugadores según su país de procedencia.
* `National_Position`: Con esta columna se implementaron los labels.

Estas variables **no debe influir** en el funcionamiento del algoritmo.

**Respuesta:**

1. Separación de datos

````python
from sklearn.model_selection import train_test_split

features = df.drop(columns = ['Name', 'Nationality', 'National_Position', 'label'])
labels = df.loc[:, 'label']

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.33, shuffle=True, stratify=labels, random_state=30
)
````

````python
# Revisar la distribución de etiquetas en los conjuntos de entrenamiento y prueba
print('Distribución de etiquetas en el conjunto de entrenamiento:')
y_train.value_counts() / y_train.count() * 100
````

````python
print('Distribución de etiquetas en el conjunto de prueba:')
y_test.value_counts() / y_test.count() * 100
````

2. Definición de *pipeline*

````python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.pipeline import Pipeline

# Definición de pipeline para entrenamiento con XGBoost
xgb_model = XGBClassifier()
pipeline_xgboost = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', xgb_model)
], verbose=True)

# Definición de pipeline para entrenamiento con LightGBM
lgb_model = LGBMClassifier(verbose=-1)
pipeline_lightgbm = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', lgb_model)
], verbose=True)
````

3. Entrenamiento del *pipeline*

````python
# Entrenamiento de XGBoost
pipeline_xgboost.fit(X_train, y_train)
````

````python
# Entrenamiento de LightGBM
pipeline_lightgbm.fit(X_train, y_train)
````

2. **Explicación de métricas**

* Accuracy: Calcula, en base a todas las observaciones, cuáles fueron clasificadas de manera correcta. Es un indicador global que agrupa todas las clasificaciones de las clases dentro del dataset. Representado como fórmula:

\begin{equation*}
    accuracy = \frac{\textit{Total correct predictions}}{\textit{Total predictions}}
\end{equation*}

Si observamos los resultados, vemos que, tanto los modelos `XGBoost` como `LightGBM` obtienen un accuracy general de $0.94$, pero este alto rendimiento es debido a que la clase mayoritaria la representan jugadores sin posición en su selección nacional, por lo que necesitamos ver métricas más específicas para la clase positiva.

* Precision: Calcula, en base a todas las observaciones categorizadas como clase positiva, la proporción de observaciones correctamente clasificadas. Representado como fórmula:

\begin{equation*}
    precision = \frac{TP}{TP + FP}
\end{equation*}

Con esta métrica obtenemos algo más realista con lo que se quiere obtener. `XGBoost` obtiene un valor $precision = 0.52$, mientras que `LightGBM` obtiene un $precision = 0.55$, siendo éste el mayor. Esto nos indica que, dentro de las observaciones catalogadas como positivas, aproximadamente un poco más de la mitad correspondían a la clase verdadera.

* Recall: Calcula, en base a todas las observaciones que debían ser clase positiva, la proporción de observaciones correctamente clasificadas. Representado como fórmula:

\begin{equation*}
    recall = \frac{TP}{TP + FN}
\end{equation*}

Por otro lado, esta métrica obtiene resultados muy bajos, ya que opera sobre las observaciones que pertenecen a la clase positiva, viendo que existe un claro sesgo hacia la clase negativa, ya que muchas observaciones positivas se predijeron erróneamente, consiguiendo $recall = 0.13$ para `XGBoost` y $recall = 0.12$ para `LightGBM`.

A modo general, ambos modelos obtuvieron excelentes métricas en relación a la clase negativa. Esto de sebe principalmente a que, como se vio anteriormente, la proporción entre clase positiva (seleccionado) y negativa (no seleccionado) es bastante dispareja. En estos mismo resultados podemos ver un total de $5.450$ observaciones de la clase negativa versus las $355$ de la clase positiva, por lo que existe un claro sesgo en el entrenamiento hacia la clase que no nos interesa predecir.

3. **Métrica más adecuada**

Para este caso, lo más recomendable sería usar la métrica *recall*, ya que estamos viendo un claro sesgo al clasificar como clase negativa sobre la positiva. * Random Forest Classifier (`rf`): Aplica la misma idea anterior, pero con múltiples árboles que trabajan de manera independiente con diferentes subconjuntos de datos. Para realizar la predicción final se hace un promedio o una votación.

A continuación, mostramos una tabla con las características más importantes de comparar en estos tres árboles.

|----------------------|-----------------------------------------------|---------------------------------------------|---------------------------------------------|

| **Subconjuntos**     | Entrenado con subconjuntos aleatorios         | Usa bootstrapping para crear árboles       | Se entrena con todos los datos disponibles  |
| **Precisión**        | Generalmente alta y menos propenso al sobreajuste | Alta, mejora la precisión y robustez     | Puede ser bajo debido al sobreajuste       |
| **Velocidad**        | Más rápido de entrenar debido a divisiones aleatorias | Más lento debido a la búsqueda de la mejor división | Rápido en entrenamiento y predicción       |
| **Interpretabilidad**| Menos interpretables debido a la complejidad del conjunto | Menos interpretables por la cantidad de árboles | Altamente interpretable y visualizable      |
| **Aplicaciones**     | Problemas donde se requiere velocidad y precisión | Tareas de clasificación y regresión variadas | Tareas donde se necesita interpretación clara |

````python
df = pd.read_csv('stats_players.csv')
df['label'] = df['National_Position'].apply(lambda x: 1 if pd.notnull(x) else 0)
````

````python
# Transformar las variables numéricas tipo calificación con MinMaxScaler
num_columns = ['Height', 'Weight', 'Age', 'Ball_Control', 'Dribbling', 'Marking', 'Sliding_Tackle', 'Standing_Tackle',
               'Aggression', 'Reactions', 'Interceptions', 'Vision', 'Composure', 'Crossing', 'Short_Pass', 'Long_Pass',
               'Acceleration', 'Speed', 'Stamina', 'Strength', 'Balance', 'Agility', 'Jumping', 'Heading', 'Shot_Power',
               'Finishing', 'Long_Shots', 'Curve', 'Freekick_Accuracy', 'Penalties', 'Volleys']
num_scaler = MinMaxScaler()

# Transformar las variables categóricas con OneHotEncoder
cat_columns = ['Name', 'Nationality', 'Club_Position', 'Preffered_Foot', 'Work_Rate', 'Weak_foot', 'Skill_Moves']
cat_encoder = OneHotEncoder(handle_unknown='infrequent_if_exist')

# Crear un ColumnTransformer para aplicar las transformaciones a las columnas correspondientes
col_transformer = ColumnTransformer([
    ('num', num_scaler, num_columns),
    ('cat', cat_encoder, cat_columns)
])
````

````python
features = df.drop(columns = ['National_Position', 'label'])
labels = df.loc[:, 'label']

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.33, shuffle=True, stratify=labels, random_state=30
)
````

````python
# Definición de pipeline para entrenamiento con XGBoost
xgb_model = XGBClassifier()
pipeline_xgboost = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', xgb_model)
], verbose=True)
````

````python
# Entrenamiento de XGBoost
pipeline_xgboost.fit(X_train, y_train)
````

````python
# Revisión de desempeño
y_pred = pipeline_xgboost.predict(X_test)
print(classification_report(y_test, y_pred))

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=pipeline_xgboost.classes_, yticklabels=pipeline_xgboost.classes_, cmap='magma')

plt.xlabel('Predicho')
plt.ylabel('Real')
plt.title('Matriz de Confusión')
plt.show()
````

**2. Clasificación de Características**

````python
import numpy as np

# Extraer los nombres de las características numéricas y codificadas
num_features = num_columns
cat_features = pipeline_xgboost.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(cat_columns)

# Combinar todos los nombres de características
all_features = np.concatenate([num_features, cat_features])

# Obtener las importancias de las características desde el modelo XGBoost
feature_importances = pipeline_xgboost.named_steps['classifier'].feature_importances_

# Crear un DataFrame para visualizar las importancias con los nombres de las características
feature_importances_df = pd.DataFrame({
    'Feature': all_features,
    'Importance': feature_importances
})

# Ordenar por importancia descendente
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# Visualizar la importancia de las primeras 100 características
plt.figure(figsize=(10, 20))
sns.barplot(data=feature_importances_df[:100], x='Importance', y='Feature', palette='magma')
plt.xlabel('Importancia')
plt.ylabel('Características')
plt.title('Importancia de las características en el modelo XGBoost')
plt.show()
````

**3. Eliminación de Características**

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Defino las características iniciales (las 500 más importantes)
features_list = feature_importances_df['Feature'].tolist()[:100]

# Copia de los datos de entrenamiento y prueba para modificar en cada iteración
X_train_iter = pd.DataFrame(X_train, columns = all_features)[features_list]
X_test_iter = pd.DataFrame(X_test, columns = all_features)[features_list]

# Iré almacenando los resultados tras cada entrenamiento
performance = []

for i in range(100, 1, -1):
    # Pipeline con las características actuales
    col_transformer_iter = ColumnTransformer([
        ('num', num_scaler, [col for col in features_list if col in num_columns]),
        ('cat', cat_encoder, [col for col in features_list if col in cat_columns])
    ])

    pipeline_iter = Pipeline([
        ('preprocessor', col_transformer_iter),
        ('classifier', XGBClassifier())
    ])

    pipeline_iter.fit(X_train_iter, y_train)

    # Evaluar el rendimiento
    y_pred = pipeline_iter.predict(X_test_iter)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label=1)
    recall = recall_score(y_test, y_pred, pos_label=1)

    performance.append({
        'Num_Features': i,
        'Accuracy': accuracy,
        'Precision (class "1")': precision,
        'Recall (class "1")': recall
    })

    # Actualizar lista de características eliminando la de menor importancia
    least_important_feature = feature_importances_df[feature_importances_df['Feature'].isin(features_list)].iloc[-1]['Feature']
    features_list.remove(least_important_feature)

    # Actualizar los datos de entrenamiento y prueba
    X_train_iter = X_train_iter[features_list]
    X_test_iter = X_test_iter[features_list]

    # Actualizar el DataFrame de importancia eliminando la característica usada
    feature_importances_df = feature_importances_df[feature_importances_df['Feature'] != least_important_feature]

performance_df = pd.DataFrame(performance)
````

1. Se solicita que utilice un método de calibración que asegure que las probabilidades generadas incrementen de manera monótona. Una métrica ampliamente utilizada para evaluar la precisión de la calibración de un modelo es el Brier Score. Calcule el Brier Score para el modelo tanto antes como después de la calibración. Esto le permitirá realizar una comparación cuantitativa y determinar si la calibración ha mejorado el rendimiento del modelo. Para más información sobre el Brier Score, puede consultar el siguiente enlace: [Scikit-Learn - Brier Score Loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html). [3 puntos]

2. Tras la calibración, examine y comente los resultados obtenidos. A su análisis añada una comparación visual de las ideales versus las salidas del modelo original (sin calibrar) y del modelo calibrado. [3 puntos]

**Respuesta**

**1. Uso de Brier Score**

````python
from sklearn.metrics import brier_score_loss

# Predecir probabilidades con el modelo no calibrado
y_prob_before = pipeline_xgboost.predict_proba(X_test)[:, 1]

# Calcular el Brier Score antes de la calibración
brier_score_before = brier_score_loss(y_test, y_prob_before)
print(f"Brier Score antes de la calibración: {brier_score_before:.4f}")

# Predecir probabilidades con el modelo calibrado
y_prob_after = pipeline_iter.predict_proba(X_test)[:, 1]

# Calcular el Brier Score después de la calibración
brier_score_after = brier_score_loss(y_test, y_prob_after)
print(f"Brier Score después de la calibración: {brier_score_after:.4f}")
````

**2. Comentarios de lo obtenido**

Como podemos ver en el resultado anterior, el valor del `brier_score_loss` tuvo una baja luego de calibrar el modelo, dando a entender que la precisión del modelo luego de calibrar tiende a la baja, ya que un menor `brier_score_loss` hubiese significado una mejora en el modelo. El que haya aumentado puede significar que el primer modelo entrenado presenta un sobreajuste en sus datos.

Para una comparación visual de las salidas ideales y originales, veamos el siguiente gráfico:

````python
from sklearn.calibration import calibration_curve

# Curvas de calibración
prob_true_orig, prob_pred_orig = calibration_curve(y_test, y_prob_before, n_bins=10)
prob_true_calib, prob_pred_calib = calibration_curve(y_test, y_prob_after, n_bins=10)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(prob_pred_orig, prob_true_orig, marker='o', label='Modelo sin Calibrar', color='blue')
plt.plot(prob_pred_calib, prob_true_calib, marker='o', label='Modelo Calibrado', color='green')

# Línea ideal
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Calibración Ideal')

plt.xlabel('Probabilidad Predicha')
plt.ylabel('Probabilidad Real')
plt.title('Curva de Calibración')
plt.legend(loc='best')
plt.grid(True)

plt.show()
````

Como se puede ver, es el modelo quien se mantiene más pegado a lo que son las predicciones reales, pero teniendo en consideración que, debido al gran desbalance de datos, estas probabilidades están mucho más referidas a la clasificación de la clase negativa sobre la positiva. La curva verde (que representa el modelo post calibración) se aleja más de de la calibración ideal, esto ya que se buscó probablemente darle más peso a las clasificaciones positivas, sin lograr tener éxito y aumentando la desviación de las predicciones.
